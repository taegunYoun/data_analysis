```python
# 0. Load Iris Dataset
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.manifold import TSNE
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns

# Load dataset
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names

df = pd.DataFrame(X, columns=iris.feature_names)
df['target'] = y
```

```python
# 1. 2D Scatter Plot (sepal length vs sepal width)
plt.figure(figsize=(8,6))
sns.scatterplot(x=iris.data[:, 0], y=iris.data[:, 1], hue=iris.target_names[y], palette='Set1')
plt.title('Iris data scatter plot')
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.legend(title='Class')
plt.grid(True)
plt.show()
```

```python
# 2. PCA 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(8,6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=iris.target_names[y], palette='Set2')
plt.title('PCA 2D visualization of iris dataset')
plt.xlabel('principal component 1')
plt.ylabel('principal component 2')
plt.legend(title='Class')
plt.grid(True)
plt.show()
```

```python
# 3. PCA 3D -  
from mpl_toolkits.mplot3d import Axes3D

pca_3d = PCA(n_components=3)
X_pca_3d = pca_3d.fit_transform(X)

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')

#  
colors = ['r', 'g', 'b']
for i, color, label in zip([0, 1, 2], colors, iris.target_names):
    ax.scatter(X_pca_3d[y == i, 0], X_pca_3d[y == i, 1], X_pca_3d[y == i, 2],
               color=color, label=label, s=60, edgecolor='k')

ax.set_title('PCA 3D visualization of iris dataset')
ax.set_xlabel('principal component 1')
ax.set_ylabel('principal component 2')
ax.set_zlabel('principal component 3')
ax.legend(title="Class")
plt.show()
```

```python
# 4. LDA 2D
lda = LDA(n_components=2)
X_lda = lda.fit_transform(X, y)

plt.figure(figsize=(8,6))
sns.scatterplot(x=X_lda[:, 0], y=X_lda[:, 1], hue=iris.target_names[y], palette='Set1')
plt.title('LDA 2D visualization of iris dataset')
plt.xlabel('LDA component 1')
plt.ylabel('LDA component 2')
plt.legend(title='Class')
plt.grid(True)
plt.show()
```

```python
# 5. t-SNE 2D -  
tsne_2d = TSNE(n_components=2, random_state=42, perplexity=30, max_iter=1000)
X_tsne_2d = tsne_2d.fit_transform(X)

plt.figure(figsize=(8,6))
sns.scatterplot(x=X_tsne_2d[:, 0], y=X_tsne_2d[:, 1], hue=iris.target_names[y], palette='Dark2')
plt.title('t-SNE 2D visualization of iris dataset')
plt.xlabel('t-SNE feature 1')
plt.ylabel('t-SNE feature 2')
plt.legend(title='Class')
plt.grid(True)
plt.show()
```

```python
# 6. t-SNE 3D -  
tsne_3d = TSNE(n_components=3, random_state=42, perplexity=30, max_iter=1000)
X_tsne_3d = tsne_3d.fit_transform(X)

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')

colors = ['r', 'g', 'b']
for i, color, label in zip([0, 1, 2], colors, iris.target_names):
    ax.scatter(X_tsne_3d[y == i, 0], X_tsne_3d[y == i, 1], X_tsne_3d[y == i, 2],
               color=color, label=label, s=60, edgecolor='k')

ax.set_title('t-SNE 3D visualization of iris dataset')
ax.set_xlabel('t-SNE feature 1')
ax.set_ylabel('t-SNE feature 2')
ax.set_zlabel('t-SNE feature 3')
ax.legend(title="Class")
plt.show()
```
